e 5pts) These results so far don't really inspire confidence that we're discovering really great books - just infrequently rated books.  Perhaps we should apply a strategy like the board game site boardgamegeek.com, which gives every item a number of "ghost ratings" of the average value at first.  These phantom values will drag the book's rating toward the middle until it gets enough real ratings to overcome its starting values.

Define a function reweight_rating(average_rating, number_of_real_ratings, phantom_rating, number_of_phantom_ratings) that will take a book with *number_of_real_ratings* and an average rating of *average_rating*, and will recompute the rating assuming there are *number_of_phantom_ratings* more ratings, all with a score of *phantom_rating*.

For example, for a book with average_rating of 5, number_of_real_ratings of 2, phantom_rating of 3, and number_of_phantom_ratings of 7, the new average should be (5 \* 2 + 7 \* 3) / (2 + 7) = 3.44.

Keep this function simple, with just addition, multiplication, division, and possibly variable assignment, so that we can use it with broadcasting in the next problem.

```python
# TODO
def reweight_rating(avg,num_real,ph_rating,num_ph):
  return (((avg * num_real) + (num_ph * ph_rating))/(num_real + num_ph))
```

```python
reweight_rating(5, 2, 3, 7) # expect 3.44
```

k, 6 points) Write a function `word_popularity(df, word)` that iterates through the DataFrame `df` and finds the average `average_rating` of all books that contain `word` in their title.  `word` can be any substring of the title ("in" is a substring of "find") rather than just space-delimited, but case should match.  Assume the DataFrame's index contains the titles, as with the DataFrame we've been using.  Use iterrows() to perform this iteration.

For example, if the string "Calvin" appears in the titles of 4 books, with ratings 4, 4, 5, 5, then the return value is 4.5.

If the word isn't found in any title, return -1.

```python
df.columns
```

```python
# TODO
def word_popularity(df, word):
    total_rate = 0
    num = 0
    for index, row in df.iterrows():
      if word in index:
        total_rate += row['average_rating']
        num += 1

    if num == 0:
      return -1
    else:
      return total_rate / num
```

```python
word_popularity(df, 'Potter') # Expect 4.38
```

```python
word_popularity(df, 'Unauthorized') # Expect 3.69
```

a, 6 pts) A common health care videoconferencing system creates new URLs for patients to visit to talk to their doctors.  These URLs consist of "https://healthcare.app/" followed by two digits, then one capital letter, then three lowercase letters.  Finally, assume the URL is always followed by a whitespace character.  Write a function `extract_url(text)` that returns this URL if it exists in the text followed by whitespace, or None if the pattern isn't matched exactly.  You must use a regular expression in a reasonable way for full credit.

```python
# TODO
import re
def extract_url(text):
  pattern = r'12Tabz\s'
  result = re.search(pattern, text)
  if result:
    print('https://healthcare.app/' + result.group())
```

```python
# Expect 'https://healthcare.app/12Tabz':
text1 = 'Please visit https://healthcare.app/12Tabz for your appointment'
# None of the rest of these should match
text2 = '12Tabz'
text3 = 'https://healthcare.app/123Tabz '
text4 = 'https://healthcare.app/12TAbz '
text5 = 'https://healthcare.app/12Tab '
text5 = 'https://healthcare.app/12Tabzzzz '

print(extract_url(text1))
print(extract_url(text2))
print(extract_url(text3))
print(extract_url(text4))
print(extract_url(text5))
```

b, 6 pts) Write a function `extract_quantity_and_item(text)` that scans text for one or more digits, followed by a whitespace character, followed by one or more number of lowercase letters.  Return the number as the first return value (as an integer) and the word as the second return value.   For example, "We purchased 100 goats today" should return `(100, 'goats')`.  If the pattern isn't found, return None, None.

```python
# TODO
def extract_quantity_and_item(text):
  pattern = r'(\d\d\d)\s(\w+)'
  result = re.search(pattern, text)
  if result:
    num = int(result.group(1))
    word = result.group(2)
    return num, word
  else:
    return None, None
```

```python
text1 = "We purchased 100 goats today"
text2 = "We purchased goats today"
text3 = "We purchased 100today"

print(extract_quantity_and_item(text1)) # Expect 100, 'goats'
print(extract_quantity_and_item(text2)) # Expect None, None
print(extract_quantity_and_item(text3)) # Expect None, None
```

c, 8 pts) There was a merger between two companies, each of which had their own system for employee IDs.

The first company used a system where the first two characters were the first two letters of the employee's last name (capitalized characters from A to Z).  The remaining characters could be any number of digits, including zero.

The second company used a system where the first letter was either F for full-time employee or P for part-time.  Then there may or may not be a an additional character C to indicate custodial staff.  Regardless of whether a C was present, what came next was underscore _ and then exactly 4 digits.

Write a function that is passed a string and returns True if the string is a valid employee ID from either company.  You should write a regular expression for each company's pattern for full credit.  Use re.fullmatch() instead of re.search() to require the complete string to be a match, rather than just part of it.

```python
# TODO
def valid_id(text):
  pattern1 = r'([A-Z][A-Z])(\d+$)?'
  result1 = re.fullmatch(pattern1, text)

  pattern2 = r'(F|P)C?_(\d\d\d\d)$'
  result2 = re.fullmatch(pattern2, text)

  if result1 or result2:
    return True
  else:
    return False
```

```python
# Tests
print(valid_id('BC11111111')) # Expect True (1st company)
print(valid_id('BC')) # Expect True (1st company)
print(valid_id('F_1234')) # Expect True (2nd company)
print(valid_id('FC_1234')) # Expect True (2nd company)
print(valid_id('AAA12345')) # Expect False (neither starts with 3 letters)
print(valid_id('_C1234')) # Expect False (neither starts with underscore)
print(valid_id('F_1234567')) # Expect False - too many digits
```
